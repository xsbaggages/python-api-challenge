```python
# Dependencies
import pandas as pd
import requests
import gmaps
import json
from scipy import stats
import pandas as pd
from config import gkey
import numpy as np
```

```python
# Configure gmaps with API key
gmaps.configure(api_key=gkey)
```

```python
weather_data = pd.read_csv("output_data/weather_data.csv", index_col=0)
```

```python
# Store 'Lat' and 'Lng' into  locations 
locations = weather_data[["Latitude", "Longitude"]].astype(float)

# Convert humidity to float and store
humidity = weather_data["Humidity"].astype(float)
```

```python
# Create a poverty Heatmap layer
fig = gmaps.figure()

heat_layer = gmaps.heatmap_layer(locations, weights=humidity, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 1)

fig.add_layer(heat_layer)

fig
```

```python
# wind speed <10
# cloudiness <10
# max temp 62 to 77 fahrenheit
ideal = weather_data[weather_data['Wind Speed']<10]
ideal = ideal[ideal['Cloudiness']<10]
ideal = ideal[(ideal['Temperature']>=62) & (ideal['Temperature']<=70)]
ideal
```

```python
# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
```

```python
# creating a list of coordinates to be used in plotting visualisations
marker_locations = []
name = []
city = []
country = []

target_radius = 5000
target_type = "hotel"
    

for index, coord in ideal.iterrows():

    # geocoordinates
    target_coordinates = str(coord['Latitude'])+","+str(coord['Longitude'])
    

    # set up a parameters dictionary
    params = {
        "location": target_coordinates,
        "keyword" : "hotel",
        "radius": target_radius,
        "type": target_type,
        "key": gkey
    }


    # run a request using our params dictionary
    response = requests.get(base_url, params=params)

    # convert response to json
    places_data = response.json()
  
    if len(places_data["results"]) != 0 :   
        
        marker_locations.append((places_data["results"][0]["geometry"]["location"]["lat"],places_data["results"][0]["geometry"]["location"]["lng"]))
        name.append(places_data["results"][0]["name"])
        
        compound = places_data["results"][0]["plus_code"]["compound_code"]

        pos = compound.index(" ")
        compound = compound[pos + 1:] 
        s = compound.split(", ")
        
        city.append(s[0])
        country.append(s[-1])

hotel_dict = {
"Hotel": name,
"City": city,
"Country": country
}
hotels = pd.DataFrame(hotel_dict)

hotels
```

```python
# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info = [info_box_template.format(**row) for index, row in hotels.iterrows()]
#locations = hotels[["Lat", "Lng"]]
```

```python

markers = gmaps.marker_layer(marker_locations, info_box_content = hotel_info)
fig = gmaps.figure()
fig.add_layer(heat_layer)
fig.add_layer(markers)
fig
```
